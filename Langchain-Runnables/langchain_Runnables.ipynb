```python
import random

class DummyLLM:

  def __init__(self):
    print('LLM created')

  def predict(self, prompt):

    response_list = [
        'My Name is Aryan Jadhao',
        'My favourite sport is Badminton',
        'I am the best AI learner in the college'
    ]

    return {'response': random.choice(response_list)}
```

```python
class DummyPromptTemplate:

  def __init__(self, template, input_variables):
    self.template = template
    self.input_variables = input_variables

  def format(self, input_dict):
    return self.template.format(**input_dict)
```

```python
template = DummyPromptTemplate(
    template = 'Write a poem about {topic}',
    input_variables = ['topic']
)
```

```python
prompt = template.format({'topic':'India'})
```

```python
llm = DummyLLM()
```

```python
llm.predict(prompt)
```

```python
class DummyLLMChain:

  def __init__(self,llm,prompt):
    self.llm = llm;
    self.prompt = prompt;

  def run(self,input_dict):
    final_prompt = prompt.format(input_dict)
    result = llm.predict(final_prompt)

    return result['response']
```

```python
template = DummyPromptTemplate(
    template = 'Write a poem about {topic}',
    input_variables = ['topic']
)
```

```python
llm = DummyLLM()
```

```python
chain = DummyLLMChain(llm,template)
```

```python
chain.run({'topic':'India'})
```

# **MENTOS**

```python
from abc import ABC, abstractmethod
```

```python
class Runnable(ABC):

  @abstractmethod
  def invoke(input_data):
    pass
```

```python
import random

class DummyLLM(Runnable):

  def __init__(self):
    print('LLM created')

  def invoke(self,prompt):
    response_list = [
        'My Name is Aryan Jadhao',
        'My favourite sport is Badminton',
        'I am the best AI learner in the college'
    ]

    return {'response': random.choice(response_list)}

  def predict(self, prompt):

    response_list = [
        'My Name is Aryan Jadhao',
        'My favourite sport is Badminton',
        'I am the best AI learner in the college'
    ]

    return {'response': random.choice(response_list)}
```

```python
llm = DummyLLM()
```

```python
class DummyPromptTemplate(Runnable):

  def __init__(self, template, input_variables):
    self.template = template
    self.input_variables = input_variables

  def invoke(self,input_dict):
    return self.template.format(**input_dict)

  def format(self, input_dict):
    return self.template.format(**input_dict)
```

```python
class RunnableConnector(Runnable):

  def __init__(self, runnable_list):
    self.runnable_list = runnable_list

  def invoke(self, input_data):
    for runnable in self.runnable_list:
      input_data = runnable.invoke(input_data)

    return input_data
```

```python
template = DummyPromptTemplate(
    template = 'Write a poem about {topic}',
    input_variables = ['topic']
)
```

```python
llm = DummyLLM()
```

```python
chain = RunnableConnector([template, llm])
```

```python
chain.invoke({'topic':'India'})
```

```python
class DummyStrOutputParser(Runnable):
  def __init__(self):
    pass

  def invoke(self,input_data):
    return input_data['response']
```

```python
parser = DummyStrOutputParser()
```

```python
chain = RunnableConnector([template, llm, parser])
```

```python
chain.invoke({'topic':'India'})
```

# **Connecting Multiple Chains**

```python
template1 = DummyPromptTemplate(
    template = 'Write a joke about {topic}',
    input_variables = ['topic']
)
```

```python
template2 = DummyPromptTemplate(
    template = 'Explain the following joke {response}',
    input_variables = ['response']
)
```

```python
llm = llm = DummyLLM()
```

```python
parser = DummyStrOutputParser()
```

```python
chain1 = RunnableConnector([template1, llm])
```

```python
chain1.invoke({'topic':'AI'})
```

```python
chain2 = RunnableConnector([template2, llm, parser])
```

```python
chain2.invoke({'response':'this is the joke'})
```

```python
final_chain = RunnableConnector([chain1, chain2])
```

```python
final_chain.invoke({'topic':'AI'})
```
